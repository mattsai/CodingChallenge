```python
from typing import List
class Solution:
    def merge(self, intervals: List[List[int]]) -> List[List[int]]:
        def get_min_max(interval_a,interval_b):
            min_a =  min(interval_a[0],interval_b[0])
            max_b =  max(interval_a[1],interval_b[1])
            return min_a,max_b

        if len(intervals) == 1 : return intervals

        intervals = sorted(intervals,key=lambda x:x[0])

        newIntervals = []
        previous = intervals.pop(0)
        newIntervals.append(previous)
        while intervals: 
            current = intervals.pop(0)
            # if len(newIntervals)>0:
            #     previous = newIntervals[-1]
               
            min_a,max_b  = get_min_max(previous,current)

            a,b = previous
            c,d = current

            if b>=c : 
                newItem  = [min_a,max_b]
                newIntervals[-1] = newItem
                previous = newItem
            else:
                newIntervals.append(current)
                previous = current
        return newIntervals
                




        
```
