```python
from typing import List
class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        min_value = float('inf') 
        min_sell = 0 
        for price in prices: 
            # if price< min_value: 
            #     min_value = price
            # else:
            min_value = min(min_value,price)
            min_sell = max(min_sell,price-min_value)
        return min_sell
        
```
