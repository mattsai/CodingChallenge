```python
class FuenteFacade:
    fuente_adapters = None
    @classmethod
    def create_font(cls,font=None,**methods):
        if cls.fuente_adapters and font:
            cls.fuente_adapters.append(SourceAdapter(font,**methods))
        else:
            #initial state
            cls.fuente_adapters = [
                SourceAdapter(FuenteA(),x_name="obtener_datos_a"),
                SourceAdapter(FuenteB(),x_name="get_data_b")
            ]
            
    @classmethod
    def run_them_all(cls):
        for source in cls.fuente_adapters: 
            source.x_name()
print('FACADE'.center(50,'*'))
FuenteFacade.create_font()
FuenteFacade.run_them_all()
print('Second time'.center(50,'-'))
FuenteFacade.create_font(FuenteC(),x_name="get_c_data")
FuenteFacade.run_them_all()
```

```python
class Observer:
    def update(self, obj, *args, **kwargs):
        raise NotImplementedError


class Observable:
    def __init__(self):
        self._observers = []

    def add_observer(self, observer):
        self._observers.append(observer)

    def remove_observer(self, observer):
        self._observers.remove(observer)

    def notify_observer(self, *args, **kwargs):
        for observer in self._observers:
            observer.update(self, *args, **kwargs)
            
class Amo(Observer): #yo xd o el maestro de todas las fuentes el que teiene que estar al pendiente
    def __init__(self,master):
        self.master = master
        
    def update(self,obj,*args,**kwargs):
        print("Master ha sido actualizado")
        print(f"Obj : {obj} arg: {args} kwargs {kwargs}")
        

        
        

class FuenteA:
    name = "soy fuente A"
    def obtener_datos_a(self):
        print('Datos de a')
        
class FuenteB:
    name = "Soy fuente B"
    def get_data_b(self) :
        print('Datos de B')
        

class FuenteFacade:
    fuente_adapters = None
    @classmethod
    def create_font(cls,font=None,**methods):
        if cls.fuente_adapters and font:
            cls.fuente_adapters.append(SourceAdapter(font,**methods))
        else:
            #initial state
            cls.fuente_adapters = [
                SourceAdapter(FuenteA(),x_name="obtener_datos_a"),
                SourceAdapter(FuenteB(),x_name="get_data_b")
            ]
            
    @classmethod
    def run_them_all(cls):
        for source in cls.fuente_adapters: 
            source.x_name()
            
    @classmethod
    def monitor_zero_source(cls,observer):
        cls.fuente_adapters[0].add_observer(observer)
            
    @classmethod
    def change_name_to_zero_source(cls,something):
        cls.fuente_adapters[0].name = something
        
            
class SourceAdapter(Observable): ##necesario que este mono derive de Observer ya que aqui se puedn cambiar atributos
    _initialised =False
    #fuente en ingles  creo idk
    def __init__(self,source,**methods): #** = kwargs
        super().__init__()
        self.source = source 
        for key,value in methods.items() : 
            func = getattr(self.source,value)
            self.__setattr__(key,func)
        self._initialised = True
        

    def __getattr__(self,attr):
        return getattr(self.source,attr)
    
    def __setattr__(self,key,value):
        if not self._initialised :
            super().__setattr__(key,value)  
        else:
            setattr(self.source,key,value)
            self.notify_observer(key=key,value=value)
            
```

```python
Narmer = Amo("Alvaro Narmer")
FuenteFacade.create_font()
FuenteFacade.run_them_all()
FuenteFacade.monitor_zero_source(Narmer)
FuenteFacade.change_name_to_zero_source("x")
```

```python
from abc import ABC, abstractmethod


class Enchufe(ABC):
    
    @abstractmethod
    def conectar(): 
        pass
    
class Telefono():
    def __init__(self,enchufe):
        self.enchufe = enchufe 
        
    def cargar(self):
        self.enchufe.conectar()

class AdaptadorUniversal(Enchufe):
    
    def __init__(self,enchufe):
        self.enchufe = enchufe
    
    def conectar(self):
        print('Adaptador Universal conectado')
        self.enchufe.conectar()
        
class EnchufeTipoC(Enchufe) :
    
    def conectar(self):
        print('He sido conectado TipoC')
    

    
```
