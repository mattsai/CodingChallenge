```python

def longestCommonPrefix( strs ) -> str:
    min_size = min(strs,key = lambda x:len(x))
    prev = ""
    cs = ""
    for i in range(len(min_size)):
        prev = ""
        for string in strs : 
            if not prev :
                prev = string[i]
                continue 
            
            if  prev == string[i]:
                prev = string[i]
                continue
            else:
                return cs
        cs += min_size[i]
            
            
    return cs
strs = ["flower","flow","flight"]
# strs = ["dog","racecar","car"]
longestCommonPrefix(strs)
```

```python

def longestCommonPrefix( strs ) -> str:
    cs = ""
    for s in zip(*strs):
        s = set(s)
        if len(s) != 1 : 
          return cs
        cs += s.pop()
    return cs
strs = ["flower","flow","flight"]
longestCommonPrefix(strs)
```
